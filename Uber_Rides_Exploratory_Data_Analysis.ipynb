#**Data Processing**

Date Column

```python
# convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

Time Column

```python
# convert Time column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

Booking ID column

```python
# removing extra quotes from Boking ID column
df['Booking ID'] = df['Booking ID'].str.replace('"', '')
```

Customer ID column

```python
# removing extra quotes from Customer ID column
df['Customer ID'] = df['Customer ID'].str.replace('"', '')
```

Avg VTAT Column

```python
# filling missing values in Avg VTAT using median
df['Avg VTAT'] = df['Avg VTAT'].fillna(df['Avg VTAT'].median())
```

Avg CTAT Column

```python
# filling missing values in Avg CTAT using median
df['Avg CTAT'] = df['Avg CTAT'].fillna(df['Avg CTAT'].median())
```

Cancelled Rides by Customer Column

```python
# replacing missing values with 0
df['Cancelled Rides by Customer'] = df['Cancelled Rides by Customer'].fillna(0)
```

Reason for cancelling by Customer Column

```python
#Missing values indicate rides that were not cancelled by customers.
#Replacing it with Not Cancelled by Customer.
df['Reason for cancelling by Customer'] = df['Reason for cancelling by Customer'].fillna('Not Cancelled by Customer')
```

Cancelled Rides by Driver Column

```python
# replacing missing values with 0
df['Cancelled Rides by Driver'] = df['Cancelled Rides by Driver'].fillna(0)
```

 Driver Cancellation Reason Column

```python
#Missing values represent rides that were not cancelled by drivers.
#Replacing it with Not Cancelled by Driver
df['Driver Cancellation Reason'] = df['Driver Cancellation Reason'].fillna('Not Cancelled by Driver')
```

Incomplete Rides Column

```python
# replacing missing values with 0
df['Incomplete Rides'] = df['Incomplete Rides'].fillna(0)
```

Incomplete Rides Reason Column

```python
#Missing values represent rides that were not incomplete.
#replacing it with
df['Incomplete Rides Reason'] = df['Incomplete Rides Reason'].fillna('Ride Completed')
```

Booking Value Column

```python
# filling missing values in Booking Value using median
df['Booking Value'] = df['Booking Value'].fillna(df['Booking Value'].median())
```

Ride Distance Column

```python
# filling missing values in Ride Distance using median
df['Ride Distance'] = df['Ride Distance'].fillna(df['Ride Distance'].median())
```

Driver Ratings Column

```python
# filling missing values in Driver Ratings using median
df['Driver Ratings'] = df['Driver Ratings'].fillna(df['Driver Ratings'].median())
```

Customer Rating Column

```python
# fill missing values in Customer Rating using median
df['Customer Rating'] = df['Customer Rating'].fillna(df['Customer Rating'].median())
```

Payment Method Column

```python
df['Payment Method'] = df['Payment Method'].fillna('Payment Information Not Avaliable')
```

Checking if it still has null values or not

#**EDA (Exploratory Data Analysis)**

```python
# How many rides were completed vs cancelled?
df['Booking Status'].value_counts().plot(kind='bar', color='teal')
plt.title("Booking Status Distribution")
plt.xlabel("Booking Status")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
# Which vehicle types are most frequently used?
df['Vehicle Type'].value_counts().plot(kind='bar', color='orange')
plt.title("Vehicle Type Usage")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Rides")
plt.show()
```

```python
# How is booking value distributed across rides?
plt.hist(df['Booking Value'], bins=30, color='green')
plt.title("Booking Value Distribution")
plt.xlabel("Booking Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# How is ride distance distributed?
plt.hist(df['Ride Distance'], bins=30, color='purple')
plt.title("Ride Distance Distribution")
plt.xlabel("Ride Distance")
plt.ylabel("Frequency")
plt.show()
```

```python
# Which payment methods are most used?
df['Payment Method'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['lavender', 'lightblue', 'lightgreen', 'pink', 'tan'])
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

```python
# Cancellations comparison
df[['Cancelled Rides by Customer','Cancelled Rides by Driver']].sum().plot(kind='bar',color=['orange', 'navy'])
plt.title("Cancellations by Customer vs Driver")
plt.xlabel("Cancellation Type")
plt.ylabel("Number of Cancellations")
plt.show()
```

```python
# Why do rides get cancelled?
df['Reason for cancelling by Customer'].value_counts().plot(kind='bar',color='orange')
plt.title("Reasons for Ride Cancellation by Customer")
plt.xlabel("Cancellation Reason")
plt.ylabel("Number of Cancellations")
plt.show()
```

```python
# What is the distribution of customer ratings?
plt.hist(df['Customer Rating'], bins=10,color="pink")
plt.title("Customer Rating Distribution")
plt.xlabel("Customer Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# At what time of day are most rides booked?
df['hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df['hour'].value_counts().sort_index().plot(kind='line')
plt.title("Bookings by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
# Which vehicle type has more cancellations?
df.groupby('Vehicle Type')['Cancelled Rides by Customer'].sum().plot(kind='bar',color="olive")
plt.title("Customer Cancellations by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Cancellations")
plt.show()
```

**Conclusion**

Time-based charts showed that ride demand varies significantly across different hours of the day and days of the week, indicating peak and off-peak periods. Frequency-based analysis highlighted how rides are distributed over time. Data cleaning steps improved accuracy and consistency, making the dataset reliable for further analysis.
